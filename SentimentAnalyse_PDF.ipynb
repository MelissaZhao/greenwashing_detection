```python
import pandas as pd
from textblob import TextBlob
```

```python
# Daten aus Excel-Datei in ein Pandas DataFrame laden
df = pd.read_excel('all.xlsx')
```

```python
# Sentiment-Analyse auf Text-Spalte durchführen und Ergebnisse in neue Spalte schreiben
df['Sentiment'] = df['TextBlocks'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df.head()
```

```python
# Speichern des Dataframes mit der neuen Spalte
df.to_parquet("data_with_sentimentscore.parquet")
```

```python
dfp = pd.read_parquet("data_with_sentimentscore.parquet")
dfp.to_excel("data_with_sentimentscore.xlsx")
```

```python
from matplotlib.pyplot import plot as plt
senti = pd.read_excel("data_with_sentimentscore.xlsx")
senti["Sentiment"].plot(kind = 'hist')
```
